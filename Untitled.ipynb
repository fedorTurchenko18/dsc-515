```python
import os
from dotenv import load_dotenv

load_dotenv()

folder_path = os.getenv('INIT_FOLDER_PATH')
folder_path

images_path = os.path.join(folder_path, 'images')
images_path

img_path = '9d/9dfccb37.jpg'
```

```python
from PIL import Image

img = Image.open(image_path)
img
```

```python

new_dir_path = os.path.join(images_path, 'phone_cases')

target_dimensions = (219, 256) 

for root, dirs, files in os.walk(new_dir_path):
    for file in files:
        # Build the full file path
        file_path = os.path.join(root, file)

        # Open the image and check its dimensions
        with Image.open(file_path) as img:
            print(img.size)
            if img.size != target_dimensions:
                print(f'Image {file} has dimensions {img.size} (expected {target_dimensions})')
```
